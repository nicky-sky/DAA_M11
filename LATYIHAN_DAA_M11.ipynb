# SOAL 1 — Hitung Jumlah Angka Genap

```python
def hitung_genap(arr):
    jumlah = 0
    for angka in arr:
        if angka % 2 == 0:
            jumlah += 1
    return jumlah

# Contoh penggunaan
data = [5, 9, 14, 16, 20]
print(hitung_genap(data))
```

1. Penjelasan pendekatan

Algoritma yang digunakan adalah iterasi linear (linear scan)

1.Inisialisasi counter bernilai 0 untuk menghitung jumlah angka genap.

2.Lakukan loop untuk setiap elemen pada array/list.

3.Untuk setiap elemen, cek kondisi: Jika angka % 2 == 0 → angka tersebut adalah genap.

4.Jika genap, tambahkan counter +1.

5.Setelah loop selesai, kembalikan total angka genap.

-Kompleksitas waktu (Time Complexity):

O(n), karena setiap elemen diproses satu kali.

-Kompleksitas memori (Space Complexity):

O(1), karena hanya menggunakan satu variabel counter.

**(Edge Cases)**

1️ Array kosong → harus mengembalikan 0
[]   → 0

2️ Array berisi angka negatif

Bilangan negatif tetap bisa genap bila habis dibagi 2.

[-2, -3, -4] → 2

3️ Array berisi angka 0

0 adalah bilangan genap (0 % 2 == 0).

[0, 1, 3] → 1

4️ Array berisi tipe selain integer

Jika input bisa campuran (misalnya string atau float), maka:

Bisa dilewati, atau

Bisa dibuat error.

Solusi aman:

if not isinstance(angka, int):
    continue

5️ Array sangat besar (ribuan/milion elemen)

Algoritma tetap efisien (O(n)), cocok untuk data besar.

#  SOAL 2 — Cari Nilai Terbesar

```python
def nilai_terbesar(arr):
    if len(arr) == 0:
        return None

    terbesar = arr[0]
    for angka in arr:
        if angka > terbesar:
            terbesar = angka
    return terbesar

# Contoh penggunaan
data = [8, 5, 2, 10]
print(nilai_terbesar(data))
```

# Cek apakah array kosong

Jika kosong, tidak ada angka untuk dibandingkan,Kita bisa mengembalikan None, atau memunculkan error seperti:

raise ValueError("Array tidak boleh kosong")


-Inisialisasi nilai awal

Ambil elemen pertama sebagai nilai awal: terbesar = arr[0]


-Iterasi setiap elemen dalam array

Bandingkan setiap elemen dengan variabel terbesar.

-Perbarui nilai jika ditemukan angka lebih besar

Jika angka > terbesar, ganti nilai terbesar dengan angka tersebut.

Setelah loop selesai, kembalikan nilai terbesar.

# SOAL 3 — Balik String

```python
def balik_string(s):
    return s[::-1]

# Contoh penggunaan
print(balik_string("besok"))  # Output: koseb
```

# Penjelasan logika yang digunakan
1. Menggunakan slicing [::-1] (cara paling Pythonic)

Notasi s[start:end:step]

Jika step = -1, maka Python mengambil karakter dari akhir ke awal.

Contoh:

"besok"[::-1] → mulai dari karakter terakhir → bergerak mundur → hasil: "koseb"

Logika:

Step negatif berarti iterasi mundur.

Python otomatis membangun string baru dengan susunan terbalik.

# SOAL 4 — Jumlahkan Semua Elemen

```python
def jumlah_elemen(arr):
    total = 0
    for x in arr:
        total += x
    return total

# Contoh
print(jumlah_elemen([3, 5, 2]))
```

# Penjelasan langkah-demi-langkah (menggunakan versi loop)

Misal input: [3, 5, 2]

Inisialisasi total = 0.

Ambil elemen pertama 3. Tambahkan ke total: total = 0 + 3 = 3.

Ambil elemen kedua 5. Tambahkan ke total: total = 3 + 5 = 8.

Ambil elemen ketiga 2. Tambahkan ke total: total = 8 + 2 = 10.

Selesai iterasi → kembalikan total yaitu 10.

# SOAL 5 — Cek Palindrome (String)

```python
def is_palindrome(s):
    s = s.lower()  # supaya tidak terpengaruh huruf kapital
    return s == s[::-1]

# Contoh penggunaan
print(is_palindrome("level"))  # True (palindrome)
print(is_palindrome("hello"))  # False (bukan palindrome)
```

# Penjelasan logika pengecekan palindrome

Cara menentukan apakah string tersebut palindrome:

Normalisasi string (opsional)

Ubah ke huruf kecil (lower()).

Jika ingin lebih ketat: hilangkan spasi dan tanda baca.

Balikkan string

Menggunakan slicing Python [::-1].

Bandingkan string asli dengan string yang sudah dibalik

Jika sama → palindrome

Jika berbeda → bukan palindrome

# SOAL 6 — Hitung Jumlah Huruf Vokal

```python
def hitung_vokal(s):
    vokal = "aiueoAIUEO"  # memasukkan huruf besar & kecil
    jumlah = 0

    for char in s:
        if char in vokal:
            jumlah += 1

    return jumlah

# Contoh penggunaan
print(hitung_vokal("informatika"))
```

# SOAL 7 — Temukan Elemen yang Dicari

```python
def cek_target(arr, target):
    for angka in arr:
        if angka == target:
            return True
    return False

# Contoh penggunaan
array = [4, 7, 1, 9]
target = 7
print(cek_target(array, target))
```
